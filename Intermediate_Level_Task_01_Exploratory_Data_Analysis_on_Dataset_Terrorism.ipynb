###Importing Libraries

```python
#Importing required Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from collections import Counter
```

###Loading Dataset

```python
#Loading Dataset
ds = pd.read_csv('globalterrorismdb_0718dist.csv', encoding ='latin1')
```

###Data Analysis

```python
ds.head()
```

###Yearly Count of Terrorist Attacks

```python
#Yearly count of terrorist attack
plt.figure(figsize=(15,10))
sns.countplot(x="iyear", data=ds)
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Per Year')
plt.show()
```

**Observation**:- 2014 records the highest number of terrorist attacks.

###Terrorist Activities by Region Every Year

```python
#Terrorist Activities By Region Every Year
pd.crosstab(ds.iyear, ds.region_txt).plot(kind='area',figsize=(20,10))
plt.title('Terrorist Activities By Region In Each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

###Region Affected by Terrorist Attacks

```python
#Region Affected by Terrorist Attack
ds['region_txt'].value_counts()
plt.figure(figsize=(15,10))
sns.countplot(x='region_txt',data=ds)
plt.xticks(rotation=90)
plt.show()
```

**Observation**:- The most affected region is Middle East & North Africa.

###Death toll yearwise

```python
#Yearwise Killed
casperyr = ds.groupby('iyear').nkill.sum().to_frame().reset_index()
casperyr.columns = ['iyear','nkill']
px.bar(data_frame=casperyr,x = 'iyear',y = 'nkill')
```

**Observation**:- It is clearly visible that 2014 has highest number of killed people.

###Analysis of Target Types

```python
#Analysis of Target types
target = list(ds['targtype1_txt'])
target_map = dict(Counter(target))
target_df = pd.DataFrame(target_map.items())
target_df.columns = ['targtype1_txt','Count']
px.bar(data_frame=target_df,x = 'targtype1_txt',y = 'Count')
```

**Observation**:- Private Citizens & Property counts is the highest among all of them.

###Analysis of Attack Type

```python
#Analysis of Attack Type
plt.subplots(figsize=(15,6))
sns.countplot(palette='inferno', x=ds['attacktype1_txt'], order=ds['attacktype1_txt'].value_counts().index)
plt.xlabel("Attack Type")
plt.xticks(rotation=90)
plt.title('Attacking Methods by Terrorists')
plt.show()
```

**Observation**:- Bombing/Explosion is the highest counts among all of the attack types.

###Top 10 countries having maximum terrorist attack

```python
#Terrorism by Countries (Top 10 Countries)
plt.subplots(figsize=(18,6))
sns.barplot(x= ds['country_txt'].value_counts()[:10].index, y=ds['country_txt'].value_counts()[:10].values, palette='viridis')
plt.title('Top Affected Countries')
plt.show()
```

**Observation**:- Iraq has the highest number of terrorist attack.

###Activity of Notorious groups

```python
#Notorious Groups
sns.barplot(y=ds['gname'].value_counts()[1:15].values,x=ds['gname'].value_counts()[1:15].index,palette='rocket')
plt.xticks(rotation=90)
plt.title('Terrorist Groups with Highest Terror Attacks')
plt.show()
```

**Observation**:- Taliban is the most notorious group according to the graph.
