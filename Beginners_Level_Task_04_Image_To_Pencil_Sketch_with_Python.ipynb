#Image To Pencil Sketch With Python

**Image Link** : https://images.news18.com/ibnlive/uploads/2021/08/national-flag-16289133273x2.jpg?impolicy=website&width=510&height=356

```python
#Importing Required Libraries
import cv2
import plotly.express as px
import matplotlib.pyplot as plt
import urllib.request
from PIL import Image
```

```python
#Loading Image
urllib.request.urlretrieve('https://images.news18.com/ibnlive/uploads/2021/08/national-flag-16289133273x2.jpg?impolicy=website&width=510&height=356',"India.jpg")
image=cv2.imread("India.jpg")
```

```python
#Converting an image from one colour space to anpother
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#Converting Image to gray scale
grayscale = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
gray = px.imshow(grayscale, color_continuous_scale='gray')
gray.show()
```

```python
#Invering the image
inv=px.imshow(255-grayscale,color_continuous_scale='gray')
inv.show()
```

```python
#Smoothing the image
G = cv2.GaussianBlur(255-grayscale,ksize=(15,15),sigmaX=0,sigmaY=0)
smoothimg=px.imshow(G,color_continuous_scale='gray')
smoothimg.show()
```

```python
#Final Image
pencilimage=cv2.divide(grayscale,255-G,scale=256)
Final_Sketch=px.imshow(pencilimage,color_continuous_scale='gray')
Final_Sketch.show()
```
